## Notion de dérivabilité en dimension 1
**Rappel** : une fonction d'un intervalle ouvert $I \subset \R$ dans $\mathbb R$ est dite *dérivable* ou *différentiable* en $x \in I$ si la quantité 
$$\lim_{h \rightarrow 0, h \neq 0} \frac{f(x+h)-f(x)}{h}$$
a une limite, qui est alors notée $f'(x)$.


Si on pose alors 
$$\epsilon(h) = \frac{f(x+h)-f(x)}{h} -f'(x)$$ la fonction $\epsilon$ 
vérifie alors
$$\lim_{h \to 0, h \neq 0} \epsilon(h) = 0.$$

On peut reformuler la notion de dérivabilité comme le fait que $f$ puisse s'écrire (pour $h$ assez petit) sous la forme
$$f(x+h) = f(x) + h \cdot f'(x) + h \cdot\epsilon(h)$$

avec $\epsilon$ une fonction qui tend vers $0$ en $0$.

La partie
$$h \mapsto f(h) + h \cdot f'(x)$$

est une approximation linéaire de $h \mapsto f(x+h)$ au voisinage de $0$ et la partie

$$h \cdot \epsilon(h) = o(h)$$

est une fonction **négligeable devant** $h$.

**exemple** Si $f(x) = x^3$ alors
$$f(x+h) = (x+h)^3 = x^3 + 3x^2 h + 3x h^2 + h^3$$
$$ = \boxed{f(x) + 3x^2 \cdot h} + h\cdot (3xh+h^2)$$
L'application $h \mapsto 3x^2 \cdot h$ est linéaire, et $h \mapsto h(3xh+h^2)$ est négligeable devant $h$ (de la forme $h \times \epsilon(h)$ avec $\lim_{h \rightarrow 0} \epsilon(h) = 0$ ).  
On retrouve donc la formule $(x^3)' = 3x^2$.

Représentation graphique des fonctions 
$$h \mapsto f(2+h) = (2+h)^3$$ 
et 
$$h \mapsto f(2) + h f'(2) = 8+12h$$

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12, 12))
h = np.linspace(-3,2,100)
plt.plot(h, (2+h)**3, color='red', label= "f(2+h)")
plt.plot(h, 8 + 12*h, color = 'green', label= "f(2) + h f'(2)")
plt.grid()
plt.axhline()
plt.axvline()
plt.legend(fontsize=20)
plt.close()
```

## L'espace ${\mathbb R}^n$
${\mathbb R}^n$ est l'espace vectoriel des $n$-uplets de réels de la forme
$$x = (x_1, x_2, \dots, x_n)$$
On peut identifier dans la base canonique ce vecteur à une matrice (vecteur colonne) de dimension $n \times 1$.
$${\bf x} = \begin{pmatrix}
x_1 \\
x_2 \\
\dots \\
x_n
\end{pmatrix}$$

On peut mesurer les distances en utilisant des **normes**. La norme d'un vecteur mesure sa "longueur" qui peut aussi être vue comme sa distance à $\bf 0$. 

Norme 2 (euclidienne) : $\|x\|_2 = \sqrt{\sum_{i=1}^n x_i^2}$  
Norme 1 : $\|x\|_1 = \sum_{i=1}^n | x_i | $    
Norme infinie : $\|x\|_\infty = \max_{1 \leq i \leq n} \| x_i \| $  

On peut alors mesurer des distances entre vecteurs par
$$d(x,y) = \|x-y\|$$  

Pour les propriétés topologiques (continuité, limites) on peut utiliser de façon *équivalente* n'importe laquelle de ces normes.

Une application linéaire de $\R^n$ dans $\R^m$ est une application 
$$ f :  \R^n \rightarrow \R^m$$
qui vérifie 
$$\forall x,y \in \R^n, f(x+y) = f(x) + f(y)$$
$$\forall \lambda \in \R, \forall x \in \R^n, f(\lambda x) = \lambda f(x)$$ 

**exemple :** $f$ de $\R^3$ dans $\R^2$ définie par

$$f(x_1,x_2,x_3) = (2x_1+x_2-x_3, x_1+3x_3)$$
est linéaire.

Une telle application peut s'écrire sous la forme
$$f(x_1, x_2, \dots, x_n) = ( \sum_{j=1}^n a_{1,j} x_j, \sum_{i=j}^n a_{2,i} x_j, \dots, \sum_{i=1}^n a_{m,j} x_j )$$

On peut représenter $f$ par une matrice à $m$ lignes et $n$ colonnes 
$$ A = (a_{i,j})_{1\leq i \leq m, 1 \leq j \leq m} $$
Les espaces $\R^n$ et $\R^m$ sont identifiés à l'ensemble des  vecteurs colonnes $n \times 1$ et $m \times 1$
et l'application linéaire correspond au produit matriciel
$$ Y = A X $$
$$ (m,1) \sim (m,n) \cdot (n,1)$$

**cas particulier** Les applications linéaires sur $\R^n$ à valeur dans $\R$ (formes linéaires) sont représentées par des matrices de dimensions $(1,n)$ donc des matrices ligne.

On peut représenter une telle ligne comme la transposée d'un vecteur colonne donc une forme linéaire
peut s'écrire sous la forme

$$f({\bf x}) = {\bf a}^T \bf x$$

où $\bf a$ est un vecteur colonne de dimension $n$. Le produit 
$${\bf a}^T {\bf x} = \sum_{j=1}^n a_i x_i$$
est aussi le produit scalaire (euclidien) des vecteurs.

**exemple :** $f$ de $\R^3$ dans $\R$
    
$$ f(x_1,x_2,x_3) = 3x_1-x_2-2x_3 $$
est représentée par le vecteur 
$${\bf a} = \begin{pmatrix}
3 \\
-1 \\
-2
\end{pmatrix}$$

## Différentielle en dimension $n$
Soit maintenant $f$ une fonction de ${\mathbb R}^n$ dans ${\mathbb R}$. La différentielle $\ell$ de $f$ en un point $x$ est une approximation linéaire de $f$ et $x$, elle doit donc réaliser
$$f(x+h) = f(x) + \ell(h) + o(h)$$
Ici $h \mapsto \ell(h)$ doit être une *application linéaire* (forme linéaire en fait) de ${\mathbb R}^n$ dans ${\mathbb R}$   
et $o(h)$ est une fonction qui est négligeable devant $h$.

Rappel : une application linéaire de  ${\mathbb R}^n$ dans ${\mathbb R}$ peut s'écrire dans une base 
sous la forme
$$\ell(h) = \sum_{i=1}^n a_i h_i$$
On peut représenter ce calcul aussi comme le produit matriciel ${\bf a}^T \cdot \bf h$ où $\bf a$ et $\bf h$ sont les vecteurs colonnes correspondant à $a$ et $h$.

**exemple :** Soit $f(x_1,x_2) = (x_1+3x_2)^2$. On considère sa différentielle au point $x =(x_1,x_2)$.  

On a $$f(x_1 + h_1, x_2 + h_2) = (x_1 + h_1 +3 x_2 + 3 h_2)^2$$
$$ = (x_1 + 3x_2)^2 + 2(x_1+3x_2)(h_1+3h_2) + (h_1+3h_2)^2$$

On peut facilement montrer que $h = (h_1,h_2) \mapsto (h_1+3h_2)^2$ est un $o(h)$ (négligeable) car
$$| (h_1+3h_2)^2 | = (h_1+3h_2)  (h_1+3h_2) \leq 3 (|h_1| + |h_2|) (h_1 + 3h_2)$$
donc est de la forme $\|h\|_1 \epsilon(h)$.

On peut donc écrire que
$$f(x+h) = f(x) +  2(x_1+3x_2)h_1 + 6(x_1+3x_2) h_2 + o(h)$$
La différentielle de $f$ en $x=(x_1,x_2)$ est l'application linéaire à valeurs dans $\R$
$$h = (h_1,h_2) \mapsto 2(x_1+3x_2)h_1 + 6(x_1+3x_2) h_2 $$
et peut être représentée par le vecteur
$${\bf a} = \begin{pmatrix}
2 x_1+ 6 x_2 \\
6 x_1 + 18x_2 
\end{pmatrix}$$

Rappel : $f(x_1,x_2) = (x_1+3x_2)^2$

On a $$\frac{\partial f}{\partial x_1}(x) = 2(x_1 + 3 x_2) = 2x_1 + 6 x_2$$
et
$$\frac{\partial f}{\partial x_2}(x) = 3 \times 2(x_1 + 3 x_2) = 6 x_1 + 18 x_2$$
On retrouve les coordonnées du vecteur de la différentielle.


Le vecteur $a$ représentant la différentielle de $f$ est appelé le **gradient de f en x** et est noté $grad(f)(x)$ ou  $\nabla f (x)$ et ses coordonnées sont 
$$\nabla f (x) = (\frac{\partial f}{\partial x_1}(x), \frac{\partial f}{\partial x_2}(x), \dots, \frac{\partial f}{\partial x_n}(x))^T$$
où $\frac{\partial f}{\partial x_i}(x)$ est la dérivée partielle de $f$ par rapport à $x_i$.

L'ensemble des vecteurs $h$ tels que $\langle \nabla f (x), h\rangle > 0$ forme un demi-espace et  ces vecteurs sont appelés directions de montée ; à l'inverse les vecteurs $h$ tels que $\langle \nabla f (x), h\rangle < 0$ sont les directions de descente.
Si h est une direction de descente alors 
$$ f(x+h) = f(x) + \langle \nabla f (x), h\rangle + o(||h||) < f(x) $$
si $h$ est assez petit ($||h|| < \epsilon$ pour un $\epsilon > 0$).

** exemple ** Toujours avec  $f(x_1,x_2) = (x_1+3x_2)^2$ et 
$${\bf \nabla f(x)} = \begin{pmatrix}
2 x_1+ 6 x_2 \\
6 x_1 + 18x_2 
\end{pmatrix}$$

En $x_0 = (2,1)$ on a $\nabla f(x) = (10, 30)^T$. Toutes les directions $h=(h_1,h_2)$ vérifiant
$$\langle \nabla f(x), h \rangle = 10 h_1 + 30 h_2 > 0$$
sont des directions de montée et donc vérifient $f(x_0+h) > f(x_0)$ pour $h$ assez petit,
et les directions vérifiant  $10 h_1 + 30 h_2 < 0$ sont des directions de descente.


## gradient et extremums
**Propriété** : Si $f$ admet un gradient en $x$ et que $x$ est un minimum ou un maximum local de $f$, alors
$$\nabla f(x) = 0$$
Attention, la réciproque est fausse. Considérer $f(x)=x^3$ en $x=0$.

Trouver les $x$ qui annulent le gradient permet trouver les candidats potentiels à être un minimum ou maximum.

*preuve* (cas du minimum) Si $\nabla f (x) \neq 0$ alors si $t$ est suffisament petit on a 
(en postant $h = - t \nabla f(x)$)
$$f(x - t \nabla f(x) )  = f(x)  -  \langle \nabla f(x), t \nabla f(x) \rangle + o(t)$$
soit
$$f(x - t \nabla f(x) )  = f(x)  -  t ||\nabla f(x)||^2 + o(t)$$
qui est $ < f(x)$ pour $t$ assez petit, contredisant le fait que $f$ est un minimum local.

Une fonction convexe sur $\R^2$

**exemple**  $f(x_1,x_2) = 2 x_1^2 -4 x_1 + 2x_1 x_2 + x_2^2$ fonction continue qui tend vers $+\infty$ quand $x,y$ tend vers l'infini donc admet (au moins) un minimum. On peut le trouver car

$$\nabla f(x_1,x_2) = (4x_1-4 +2x_2, 2x_1 + 2x_2)^T = \bf 0$$

seulement si $x = (2,-2) $

Comme $f$ doit avoir un minimum c'est forcément $(2,-2)$.

```python
import numpy as np
a1 = np.arange(0,4,0.05)
b1 = np.arange(-4,0,0.05)
A, B = np.meshgrid(a1,b1)
```

```python
import matplotlib.pyplot as plt
f = lambda x1,x2: 2*x1**2 - 4*x1 +2*x1*x2 + x2**2
Z = f(A,B) ; levels = [-4+i*0.1 for i in range(50)]
plt.figure(figsize=(8,8))
cp = plt.contourf(A, B, Z, levels)
plt.colorbar(cp) ; plt.show()
```

## Descente de gradient
### algorithme
Bien souvent (en grande dimension) il n'est pas possible de résoudre le système d'équations $\nabla f(x)=\bf 0$.
On utilise alors une méthode *itérative*.  
La *descente de gradient* est un algorithme pour calculer un minimum d'une fonction.
L'algorithme produit une suite de points (vecteurs) $p_0, p_1, p_2, \dots$ et va s'arrêter quand le gradient de $f$
est suffisament petit ( $\| \nabla f\| <  \epsilon$ )

* paramètres : choisir $\epsilon >0$ la tolérance, et $\alpha > 0$ le pas de gradient
* Initialisation : choisir $p_0$, poser $k =0$
* Tant que $\| \nabla f(p_k)\| \geq \epsilon$ répéter :
    * $\displaystyle p_{k+1} = p_k - \alpha \nabla f(p_k)$ 
    * $k \leftarrow k+1$

Dans le cas général il n'y a pas de garantie de trouver un minimum par cette manière. Cependant si $f$ est convexe (forme en bol comme $x^2$ ) alors la méthode convergera vers l'unique minimum de $f$. Il peut être nécessaire pour cela de faire diminuer le pas lorsqu'on approche du minimum.

### Exemple
On considère la fonction sur ${\mathbb R}^3$
$$f(x_1,x_2,x_3) = 7x_1^2 - 6 x_1 x_2 + 5 x_2^2 - 12x_1 x_3 + 8 x_2 x_3 + 6 x_3^2 - 36x_1 + 44x_2 + 40x_3 + 100$$


```python
f = lambda x1,x2,x3 : 7*x1**2 - 6*x1*x2 + 5*x2**2 - 12*x1*x3 + 8*x2*x3 + 6*x3**2 - 36*x1 + 44*x2 + 40*x3 + 100
diff = lambda x1,x2,x3 : np.array((14*x1 - 6*x2 - 12*x3 - 36, -6*x1 + 10*x2 + 8*x3 + 44, -12*x1 + 8*x2 + 12*x3 + 40))
```

```python
import numpy as np
def descente_gradient(f, diff, epsilon, alpha, p0):
    p = p0
    k = 0
    points = [p0]
    while (diff(*p)**2).sum() >= epsilon**2 and k<10000:
        p = p - alpha*diff(*p)
        print( "f(" +str(p)+ ")=" + str(f(*p)))
        k +=1
        points.append(p)
    return k, np.array(points)
        
```

```python
descente_gradient(f, diff, 0.01, 0.06, np.array((20.0,20.0,20.0)))
```

**Etude  théorique**
On a dans la base canonique
$$\nabla f (x) = (14x_1 - 6x_2 - 12x_3 - 36, -6x_1 + 10x_2 + 8x_3 + 44, -12x_1 + 8x_2 + 12x_3 + 40)$$
Résoudre $\nabla f (x) =0$ donne une solution unique $x^* = (3,-5,3)$ et $f(x^*)= -4$.

### Comportements de la descente de gradient
Considérons la fonction $f(x)=x^2$ ayant un unique minimum en $x=0$. Comme $f'(x)=2x$ la descente de gradient consiste à poser 
$$x_{k+1} = x_k - \alpha \cdot 2x_k = (1-2\alpha)x_k$$
On voit essentiellement deux cas :
*  Si $ \alpha \geq 1$ alors  $|1-2\alpha| \geq 1$ oscille entre $x_k$ et $-x_k$ (pour $\alpha=1$) ou bien alterne entre positif et négatif avec $|x_k| \rightarrow \infty$ (pour $\alpha > 1)$
*  Si $0 < \alpha < 1$ alors $x_k$ va converger vers $0$:  
    * en une étape si $\alpha = \frac1{2}$  
    * sinon le temps pour obtenir $| x_k | < \epsilon$ est 
        $$\frac{ \log( x_0 / \epsilon)}{ - \log | 1 - 2\alpha |  }$$

```python
# TEMPS DE CONVERGENCE EN FONCTION DE ALPHA
X = np.linspace(0.01, .49,100)
plt.plot(X, -1/np.log(1-2*X))
plt.show()
```

```python
# GRADIENT POUR ALPHA PETIT : convergence lente
f = lambda x : x**2
diff = lambda x: 2*x
k, points = descente_gradient(f, diff, epsilon = 0.01, alpha = 0.05, p0 = np.array([5]))
```

```python
X = np.linspace(-5,5,200)
plt.plot()
Z = points[0:10]
plt.plot(X,X**2, 'r')
plt.scatter(Z,Z**2)
plt.show()
```

```python
# GRADIENT POUR ALPHA = environ 0.5 : convergence  très rapide
k, points = descente_gradient(f, diff, epsilon = 0.01, alpha = 0.45, p0 = np.array([5]))
```

```python
# GRADIENT POUR ALPHA TROP GRAND : pas de convergence
k, points = descente_gradient(f, diff, epsilon = 0.01, alpha = 1.2, p0 = np.array([5]))
```

```python
X = np.linspace(-100,100,200)
plt.plot()
Z = points[0:10]
plt.plot(X,X**2, 'r')
plt.scatter(Z,Z**2)
plt.show()
```

**Problème du réglage du pas de gradient $\alpha$** :
  * si $\alpha$ est trop grand l'algorithme ne converge pas
  * si $\alpha$ est trop petit l'algorithme converge mais trop lentement
 Il existe donc une façon optimale de régler $\alpha$ entre ces deux extremes qu'il faut trouver.  
 Il y a aussi des méthodes consistant à diminuer $\alpha$ au fur et à mesure.
