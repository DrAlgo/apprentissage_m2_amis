# Tutoriel pyplot

## importer numpy et pyplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

## courbe simple reliant des points

On veut représenter les points (1,3)   (3,5) et (7,-1)
Il faut définir un tableau X pour les abscisses et un tableau de même taille Y pour les ordonnées.

```python
X = np.array([1,3,7])
Y = np.array([3,5,-1])
```

```python
plt.plot(X,Y)  #instruction de base pour faire une courbe
plt.show()     #à ajouter en fin de plot pour qu'il apparaisse
```

Ajoutons quelques options : couleur, ligne pointillée, légende des axes, titre

```python
plt.plot(X,Y, color = 'pink', linestyle='--', linewidth=5)   #quelques options sur le style
plt.xlabel('les abscisses')
plt.ylabel('les ordonnées')
plt.title('une courbe')
plt.show() 
```

## Ajout d'une courbe de fonction

On va ajouter sur le premier graphique la courbe de la fonction $x^2-3*x$ pour $x$ variant de 1 à 7. Pour cela il faut définir de nombreux points régulièrement entre 1 et 7, on en prendra 100 ici (voir Xlin) et on notera Yf l'image de ces points par la fonction.

```python
Xlin = np.linspace(1,7,100)   # 100 points de 1 à 7 (on peut aussi utiliser np.arange)
f = lambda x:x**2-3*x         # définition de la fonction (on peut aussi utiliser def)
Yf = f(Xlin)                  # la fonction s'applique à chaque entrée du vecteur
```

```python
plt.figure(figsize=(8,8))   # pour changer la taille de la figure
plt.plot(X,Y, color = 'pink', linestyle='--', linewidth=5, label='première courbe') 
plt.plot(Xlin,Yf, color = 'green', label='deuxième courbe')
plt.xlabel('les abscisses')
plt.ylabel('les ordonnées')
plt.title('une courbe')
plt.legend()  # pour afficher les labels
plt.show() 
```

## Affichage de points isolés

Si on veut maintenant afficher les mêmes points X et Y sans les relier on utilise la commande *scatter*

```python
plt.scatter(X,Y)
plt.grid()       # pour afficher la grille
plt.axhline()    # axe des x
plt.axvline()    # axe des y
plt.show()
```

### obtenir des points régulièrement espacés dans le plan

On va cherher à obtenir des points dans le carré $x$ de -1 à 3 et $y$ de 2 à 5

```python
X1 = np.linspace(-1,3,20)  #X1 de longueur 20 
Y1 = np.linspace(2,5,20)   #Y1 de longueur 20
print(X1)
```

On veut maintenant créer la liste des tous les points $(x,y)$ avec $x \in X1$ et $y \in Y1$ mais pour le représenter avec pyplot il faut la liste de tous les abscisses de ces points puis la liste de tous leurs ordonnées.
Ceci se fait avec la commande *meshgrid*.

```python
X2, Y2 = np.meshgrid(X1,Y1)
```

```python
plt.scatter(X2,Y2)
plt.show()
```

## Des points colorés suivant la valeur d'une fonction

On va maintenant changer la couleur en fonction de la valeur de la fonction de deux variables
$$x^2 - 5*y$$

```python
Z2 = X2**2 - 5*Y2  #comme c'est simple, pas obligés de définir la fonction à part
plt.scatter(X2,Y2, c=Z2, cmap='plasma')
plt.show()
```

**cmap** est la *color mp* qui va associer une couleur à chaque valeur de Z. De nombreux choix sont possibles, voir https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html

```python
# si on refait ça avec plus de points on verra un degradé sans les points
X3 = np.linspace(-1,3,100)
Y3 = np.linspace(2,5,100) 
X4, Y4 = np.meshgrid(X3,Y3)
Z4 = X4**2 - 5*Y4  
cp = plt.scatter(X4,Y4, c=Z4, cmap='inferno')
plt.colorbar(cp)  # pratique pour voir les valeurs
plt.show()
```

## lignes de niveau d'une fonction
Il s'agit de la même chose mais on colorie par zones, nous donnons une liste de valeurs limites à tracer (*levels*)

```python
levels = np.linspace(-25,0,10)   #on se base sur les valeurs vues ci dessus
cp = plt.contourf(X4, Y4, Z4, levels, cmap='inferno')
plt.colorbar(cp) 
plt.show()
```
